### 1. How is the gender composition distributed among employees within the company? 
    -- The gender distribution is calculated by counting the occurrences of "Male" and "Female" values in the gender column of the HR Database.

```python
%%sql
SELECT gender, COUNT(*) AS count
FROM hr
WHERE age >= 17
GROUP BY gender
```

### 2. Could you provide an overview of the distribution of race and ethnicity among the company's workforce?

    -- The race/ethnicity breakdown is obtained by counting the occurrences of different race/ethnicity values in the race column of the HR Database Table.

```python
%%sql
SELECT race, COUNT(*) AS count
FROM hr
WHERE age >= 17
GROUP BY race
ORDER BY count DESC
```

### 3. Can you share insights regarding the distribution of employee ages across the company?
    -- The age distribution is determined by calculating the range of ages based on birthdate information and counting the occurrences within each age range.


```python
%%sql
SELECT 
  MIN(age) AS youngest,
  MAX(age) AS oldest
FROM hr
WHERE age >= 17
```

```python
%%sql
SELECT FLOOR(age/10)*10 AS age_group, COUNT(*) AS count
FROM hr
WHERE age >= 17
GROUP BY FLOOR(age/10)*10;
```

### Using class interval to group the age of employee, to know the total number of employee's who falls in a certain age Group

```python
%%sql
SELECT 
  CASE 
    WHEN age >= 17 AND age <= 24 THEN '17-24'
    WHEN age >= 25 AND age <= 34 THEN '25-34'
    WHEN age >= 35 AND age <= 44 THEN '35-44'
    WHEN age >= 45 AND age <= 54 THEN '45-54'
    WHEN age >= 55 AND age <= 64 THEN '55-64'
    ELSE '65+' 
  END AS age_group, 
  COUNT(*) AS count
FROM 
  hr
WHERE 
  age >= 17
GROUP BY age_group
ORDER BY age_group;
```

```python
%%sql
SELECT 
  CASE 
    WHEN age >= 17 AND age <= 24 THEN '17-24'
    WHEN age >= 25 AND age <= 34 THEN '25-34'
    WHEN age >= 35 AND age <= 44 THEN '35-44'
    WHEN age >= 45 AND age <= 54 THEN '45-54'
    WHEN age >= 55 AND age <= 64 THEN '55-64'
    ELSE '65+' 
  END AS age_group, gender,
  COUNT(*) AS count
FROM 
  hr
WHERE 
  age >= 17
GROUP BY age_group, gender
ORDER BY age_group, gender;
```

### 4. What is the division between employees working at the company's headquarters and those based in remote locations?

    -- The age distribution is determined by calculating the range of ages based on birthdate information and counting the occurrences within each age range.

```python
%%sql
SELECT location, COUNT(*) as count
FROM hr
WHERE age >= 17
GROUP BY location;
```

### 5. Could you offer information about the average duration of employment for employees whose tenure has concluded?

    --The average length of employment for terminated employees is calculated by determining the difference between the termination date and hire date for each terminated employee, and then calculating the average of these differences.

```python
%%sql
SELECT ROUND(AVG(DATEDIFF(termdate, hire_date))/365,0) AS avg_length_of_employment
FROM hr
WHERE termdate <> '0000-00-00' AND termdate <= CURDATE() AND age >= 17;
```

### 6. How do gender proportions differ across various departments and job titles?
    -- Gender distribution across departments and job titles is achieved by grouping the data by department, job title, and gender, and then counting the occurrences in each group.

```python
%%sql 
SELECT department, gender, COUNT(*) as count
FROM hr
WHERE age >= 17
GROUP BY department, gender
ORDER BY department;
```

### 7. Can you outline the prevalence of different job titles across the entire company?
    -- The distribution of job titles is obtained by grouping the data by job title and counting the occurrences of each job title.

```python
%%sql
SELECT jobtitle, COUNT(*) as count
FROM hr
WHERE age >= 17
GROUP BY jobtitle
ORDER BY jobtitle DESC;
```

### 8. Which department exhibits the highest rate of turnover among its employees?

    -- "Turnover rate" commonly refers to the frequency at which employees exit a company or department, necessitating replacements. It can be computed by dividing the count of departing employees during a specific timeframe by the average number of employees present in the company or department within that corresponding timeframe.

```python
%%sql
SELECT department, COUNT(*) as total_count, 
    SUM(CASE WHEN termdate <= CURDATE() AND termdate <> '0000-00-00' THEN 1 ELSE 0 END) as terminated_count, 
    SUM(CASE WHEN termdate = '0000-00-00' THEN 1 ELSE 0 END) as active_count,
    (SUM(CASE WHEN termdate <= CURDATE() THEN 1 ELSE 0 END) / COUNT(*)) as termination_rate
FROM hr
WHERE age >= 17
GROUP BY department
ORDER BY termination_rate DESC;
```

### 9.What is the dispersion of employees across various city and state locations?
    -- The distribution of employees across different states is calculated by grouping the data by location state and counting the occurrences in each group.

```python
%%sql
SELECT location_state, COUNT(*) as count
FROM hr
WHERE age >= 17
GROUP BY location_state
ORDER BY count DESC;
```

### Can you determine the historical shifts in the company's employee count over time, considering both hires and terminations?

    -- **This query groups the employees by the year of their hire date and calculates the total number of hires, terminations, and net change (the difference between hires and terminations) for each year. The results are sorted by year in ascending order.**

```python
%%sql
SELECT 
    YEAR(hire_date) AS year, 
    COUNT(*) AS hires, 
    SUM(CASE WHEN termdate <> '0000-00-00' AND termdate <= CURDATE() THEN 1 ELSE 0 END) AS terminations, 
    COUNT(*) - SUM(CASE WHEN termdate <> '0000-00-00' AND termdate <= CURDATE() THEN 1 ELSE 0 END) AS net_change,
    ROUND(((COUNT(*) - SUM(CASE WHEN termdate <> '0000-00-00' AND termdate <= CURDATE() THEN 1 ELSE 0 END)) / COUNT(*) * 100),2) AS net_change_percent
FROM 
    hr
WHERE age >= 17
GROUP BY 
    YEAR(hire_date)
ORDER BY 
    YEAR(hire_date) ASC;
```
